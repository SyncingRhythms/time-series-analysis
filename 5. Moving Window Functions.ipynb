```python
%matplotlib inline 
import matplotlib.pylab
import numpy as np
import pandas as pd
```

```python
# Window functions are like aggregation functions
# You can use them in conjunction with .resample() which will soon be a deferred operation
```

```python
df = pd.DataFrame(np.random.randn(600, 3), index = pd.date_range('7/1/2016', freq = 'S', periods = 600), columns = ['A', 'B', 'C'])
```

```python
#pd.rolling_mean(df, window = 2)[1:10] # in future versions you want to resample separately
r = df.rolling(window = 10)
#r.agg, r.apply, r.count, r.exclusions, r.max, r.median, r.name, r.quantile, r.kurt, r.cov, r.corr, r.aggregate, r.std, r.skew, r.sum, r.var
df.plot(style = 'k--')
r.mean().plot(style = 'k')
r.quantile(.5).plot(style = 'k')
```

```python
#exponentially weighted
#pd.ewma(df, span = 15, min_periods = 2)
df.ewm(span=15,min_periods=2,adjust=True,ignore_na=False).mean()
```

```python
# can also easily plot a rolling average per column
df = pd.DataFrame(np.random.randn(1000, 4), index = pd.date_range('6/6/16', periods = 1000), columns = ['A', 'B', 'C', 'D'])
```

```python
df.head()
```

```python
# What about a custom function?
df.rolling(window = 10).apply(lambda x: np.fabs(x - x.mean()).mean())[:15]
```

```python
# What about a custom function?
df.rolling(window = 10).apply(lambda x: x[0] - x[-1])
```

```python
# Expanding windows
# Yields the value of the statistic with all the data available up to that point in time
df.expanding(min_periods = 1).mean()[1:5]
```

```python
# Will an expanding window provide a more or less stable statistic than a rolling statistic? 
# When might one vs. the other be appropriate?
```

* expanding is more stable because it includes all previous timepoints, and thus, changes slowly over time. And, it levels out pretty quickly.
* rolling is preferred when accurate representation of more recent timepoints is preferred

```python
# EWMA
# Why use EWMA?
# 
```

* exponentially weighted mean window - to weight points in the center of the window more heavily, provides a smooth dropoff as you move to the edges of your window
